Importing all the modules required

```python
import tensorflow 
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import sklearn
from sklearn.model_selection import train_test_split
import numpy as np
import pandas as pd
from sklearn.metrics import mean_squared_error
```

Reading the data using read_csv method

```python
a=pd.read_csv('concrete_data.csv')
a.head()
```

Assigning the values except the strength column to x 

```python
x=a[['Cement','Blast Furnace Slag','Fly Ash','Water','Superplasticizer','Coarse Aggregate','Fine Aggregate']]

x.head()
```

Assigning the values of strength column to x 

```python
y=a[['Strength']]
```

PART - A

1.  Converting x and y into an numpy array

2.  Splitting 30 % of data into test dataset and the rest to train dataset

3.  Creating a model with 7 input ,one hidden layer of 10 nodes and  with a relu activation function and a output layer

4.  Initialising the model with adam optimizer and a loss function of mean squared error 

5.  Creating a empty list for the getting the 50 mean square error values in 50 iterations with epoch value as 50

6.  Converting c list to numpy array

```python
x=np.array(x)
y=np.array(y)


x1,x2,y1,y2=train_test_split(x,y,test_size=0.3)


model0=Sequential()
model0.add(Dense(10,activation='relu',input_shape=(7,)))
model0.add(Dense(1))


model0.compile(optimizer='adam',loss='mean_squared_error')


c=[]
for i in range(50):
    model0.fit(x1,y1,epochs=50,verbose=2,validation_split=0.3)
    x4=model0.predict(x2)
    k=mean_squared_error(y2,x4)
    c.append(k)
    
    
c=np.array(c)
print('mean ='+str(c.mean()))
print('standard deviation ='+str(c.std()))
```

PART - B

1.  Converting x and y into an numpy array and converting x to a normalized form of data

2.  Splitting 30 % of data into test dataset and the rest to train dataset

3.  Creating a model with 7 input ,one hidden layer of 10 nodes and  with a relu activation function and a output layer

4.  Initialising the model with adam optimizer and a loss function of mean squared error

```python
x=np.array(x)
y=np.array(y)


x=(x-x.mean())/x.std()


x1,x2,y1,y2=train_test_split(x,y,test_size=0.3)


model1=Sequential()
model1.add(Dense(10,activation='relu',input_shape=(7,)))
model1.add(Dense(1))


model1.compile(optimizer='adam',loss='mean_squared_error')


c=[]
for i in range(50):
    model1.fit(x1,y1,epochs=50,verbose=2,validation_split=0.3)
    x4=model1.predict(x2)
    k=mean_squared_error(y2,x4)
    c.append(k)
    
    
c=np.array(c)
print('mean ='+str(c.mean()))
print('standard deviation ='+str(c.std()))
```
